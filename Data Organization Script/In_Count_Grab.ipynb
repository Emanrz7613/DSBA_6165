# Grabbing images with 0-6 sperm cell count
### This notebook grabs teh images with a 1-6 sperm cell count utilizing the sperm_counts_per_frame.csv It will also combine the sperm_counts and BBs into one file and save those with the images in a new folder

```python
import pandas as pd
import numpy as np
import os as os
import shutil
```

```python
#Change to your directory
counts = pd.read_csv("C:/Users/Derrick Mayall/GitHub_Repos/6165/Data/Counts1-6 Data/sperm_counts_per_frame.csv")
bb = pd.read_csv("C:/Users/Derrick Mayall/GitHub_Repos/6165/Data/sperm_all_BBs.csv", sep=' ')
counts
```

```python
#getting all instances in range 0-10
counts = counts[counts[' sperm_count'].isin(range(0,7))]
counts
```

```python
#making columns to make later processes work faster such as grabbing folder name and frame numbers
counts['folder'] = [i[:2] for i in counts['frame_name']]
counts
```

```python
#Creating new directory and copyting the data in this df to it
new_directory = "C:/Users/Derrick Mayall/GitHub_Repos/6165/Data/Train_test/"

if not os.path.exists(new_directory):
    os.mkdir(new_directory)
```

```python
#looping through folders and copying folders over
directory = "C:/Users/Derrick Mayall/GitHub_Repos/6165/Data/Counts1-6 Data/"

for i in counts['folder'].unique():
    temp = counts[counts['folder'] == i]
    for j in temp['frame_name'].unique():
        file = directory + i + '/images/' + j + '.jpg'
        destination = new_directory + '/' + j + '.jpg'
        shutil.copy( file, destination)
```
